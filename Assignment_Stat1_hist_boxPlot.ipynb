# **`***Box***`** **Plot**

```python
# Boxplot
fd=plt.figure(figsize=(10,9))
plt.boxplot(df['Volume'],fd)
plt.title('Boxplot of Volume')
plt.xlabel('Volume')
plt.ylabel('Volume')
plt.show()
```

```python
import numpy as np

Q1 = np.percentile(df['Volume'], 25)
Q3 = np.percentile(df['Volume'], 75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = [x for x in df['Volume'] if x < lower_bound or x > upper_bound]
print(f"Identified outliers: {outliers}")
```
