```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules

# Sample data

data = pd.read_csv('preprocessed_transactions.csv', header=None)
data.columns=['Transactions']
# Split transactions into lists of items
data['Transactions'] = data['Transactions'].apply(lambda x: x.split(', '))

# Display the transformed DataFrame
#print(data)
# Convert transactions to list of lists format
transaction_list = data['Transactions']
# Encode the transaction data
te = TransactionEncoder()
te_ary = te.fit(transaction_list).transform(transaction_list)
df = pd.DataFrame(te_ary, columns=te.columns_)
df
```

```python
# Find frequent itemsets with minimum support threshold
min_support = 0.01  # Adjust the support threshold as needed
frequent_itemsets = apriori(df, min_support=min_support, use_colnames=True)

# Print frequent itemsets
print(frequent_itemsets)
```

```python
# Generate association rules
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)

# Print the generated rules
print(rules)
```

# Analysis and Interpretation:

```python
# Example filtering rules (adjust thresholds as needed)
filtered_rules = rules[(rules['support'] > 0.1) & (rules['confidence'] > 0.5) & (rules['lift'] > 1.2)]
```

```python
sorted_rules = rules.sort_values(by='lift', ascending=False)
```


A lift of 2.93 for the combination of "herb & pepper" and "ground beef" signifies a strong positive association. Understanding such associations through lift analysis helps retailers and marketers make informed decisions to improve sales, customer experience, and operational efficiency.

A confidence of 0.26 for "herb & pepper" and "ground beef" indicates a moderate association where approximately 26% of transactions containing "herb & pepper" also contain "ground beef". This insight can guide retail strategies to enhance product offerings, promotions, and customer experience based on observed purchasing patterns.
